# How to plot a bar chart with python

Bảng dưới đây cho thấy độ tuổi của những người nghiện ma túy tại một trung tâm cai nghiện vào các năm 1971, 1972.

|        ages      |1971  |1972  |
|------------------|------|------|
|   dưới 20        |118   |96    |
|   Từ 20 - dưới 25|772   |727   |
|   Từ 25 - dưới 30|288   |376   |
|   Từ 30 - dưới 35|112   |117   |
|   Từ 35 - dưới 50|112   |118   |
|   Trên 50        |117   |165   |
<ol type="a">
  <li>Vẽ biểu đồ thanh (bar chart) cho năm 1972</li>
  <li>Vẽ biểu đồ so sánh 2 năm.</li>
<ol>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

age = ["Dưới 20", "Từ 20 - dưới 25", "Từ 25 - dưới 30", "Từ 30 - dưới 35", "Từ 35 - dưới 50", "Trên 50"]
num_of_people_1971 = [118, 772, 288, 112, 112, 117]
num_of_people_1972 = [96, 727, 376, 117, 118, 165]

# Các chỉ số (index) cho các thanh (bar)
ind = np.arange(len(age))

# Độ lớn của thanh 
width = 0.35

# Vẽ
fig, ax = plt.subplots(figsize=(12, 6))

# Vẽ biểu đồ thanh ngang
ax.barh(age, num_of_people_1972)
ax.set_xlabel('Số người', fontweight = 'bold')
ax.set_ylabel('Độ tuổi', fontweight = 'bold')
ax.set_yticks(ind)
ax.set_yticklabels(age)

# Hàm dán nhãn cho các thanh
def add_value_labels(ax, spacing=5):  
    # Vởi mỗi thanh sẽ đặt một nhãn
    for rect in ax.patches:
        # Xác định vị trí X, Y để đặt các nhãn
        x_value = rect.get_width()
        y_value = rect.get_y() + rect.get_height()/4

        # Khoảng cách giữa thanh và nhãn
        space = spacing

        # Dùng x_value để làm nhãn cho các thanh
        label = "{:}".format(x_value)

        # Tạo chú thích
        ax.annotate(
            label,                      # Dùng label làm nhãn
            (x_value, y_value),         # Đặt label ở bên trái các thanh
            xytext=(space, 0),          # Đặt lable cách thanh một khoảng bằng space
            textcoords="offset points") # Interpret `xytext` as offset in points

# Call the function above. All the magic happens there.
add_value_labels(ax)

ax.set_title(
    'Số người nghiện ma túy tại một trung tâm cai nghiện vào năm 1972', 
    fontsize = 15,
    fontweight = 'bold',
    pad = 20,
    loc = 'center')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
width = 0.35
plot1 = ax.barh(ind - width/2, num_of_people_1971, width, label='Năm 1971')
plot2 = ax.barh(ind + width/2, num_of_people_1972, width, label='Năm 1972')
add_value_labels(ax)
ax.set_xlabel('Số người', fontweight = 'bold')
ax.set_ylabel('Độ tuổi', fontweight = 'bold')
ax.set_yticks(ind)
ax.set_yticklabels(age)
ax.set_title(
    'Số người nghiện ma túy tại một trung tâm cai nghiện vào năm 1971 và 1972',
    fontsize = 15,
    fontweight = 'bold',
    pad = 20,
    loc = 'center')
plt.legend()
plt.show()
```
