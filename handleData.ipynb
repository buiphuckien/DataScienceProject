# How to plot a pie chart with python

Bảng dưới đây cho thấy độ tuổi của những người nghiện ma túy tại một trung tâm cai nghiện vào các năm 1971, 1972.

|        ages      |1971  |1972  |
|------------------|------|------|
|   dưới 20        |118   |96    |
|   Từ 20 - dưới 25|772   |727   |
|   Từ 25 - dưới 30|288   |376   |
|   Từ 30 - dưới 35|112   |117   |
|   Từ 35 - dưới 50|112   |118   |
|   Trên 50        |117   |165   |

Vẽ biểu đồ hình quạt (pie chart) cho năm 1971.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math

age = ["Dưới 20", "Từ 20 - dưới 25", "Từ 25 - dưới 30", "Từ 30 - dưới 35", "Từ 35 - dưới 50", "Trên 50"]
num_of_people_1971 = [118, 772, 288, 112, 112, 117]

fig, ax = plt.subplots(figsize=(12, 6))

def func(pct, num_of_people):
    absolute = int(np.round(pct/100.*np.sum(num_of_people_1971)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

wedges, texts, autotexts = ax.pie(num_of_people_1971, autopct=lambda pct: func(pct, num_of_people_1971),
                                  textprops=dict(color="w"))

ax.legend(wedges, age,
          title="Độ tuổi",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10)

ax.set_title(
    "Số người nghiện ma túy tại một trung tâm cai nghiện vào năm 1971",
    fontsize=15,
    fontweight = 'bold',
    loc = 'center')
plt.show()
```
